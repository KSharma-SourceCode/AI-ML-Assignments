# Exploratory Data Analysis (EDA)

## Objective: Understand data distribution, feature relationships, and target balance before model development.

## 1. Class Balance Visualization :

### The dataset shows a relatively balanced distribution between patients with and without heart disease. Therefore, no additional class imbalance handling techniques (such as SMOTE) are required at this stage.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/heart_clean.csv")
df.head()

plt.figure(figsize=(6,4))
sns.countplot(x="target", data=df)
plt.title("Class Balance: Heart Disease Presence")
plt.xlabel("Target (0 = No Disease, 1 = Disease)")
plt.ylabel("Number of Samples")
plt.show()
```

## 2. Feature Distribution â€“ Histograms :

### Most numerical features such as age, resting blood pressure, and cholesterol follow approximately normal distributions, while features like fasting blood sugar and exercise-induced angina are skewed. This motivates feature scaling during model training.

```python
df.hist(figsize=(14,10), bins=20)
plt.suptitle("Feature Distributions (Histograms)", fontsize=16)
plt.show()
```

## 3. Correlation Heatmap

### Chest pain type, maximum heart rate achieved, and exercise-induced angina show strong correlation with the target variable. No extreme multicollinearity is observed, making the dataset suitable for classical machine learning models.

```python
plt.figure(figsize=(12,8))
sns.heatmap(
    df.corr(),
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Heatmap of Features")
plt.show()
```

# EDA Summary

- The dataset is clean with minimal missing values  
- Target classes are reasonably balanced  
- Feature distributions suggest the need for scaling  
- Several clinical features show meaningful correlation with heart disease presence  
- These insights guide preprocessing and model selection in subsequent stages
